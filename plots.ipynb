```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.loadtxt('results/phi_init.dat')
x=A[:,0]
f=A[:,1]

phi = np.loadtxt('results/phi.dat')

plt.figure()
plt.plot(x,f**2,'k',label='phi_init')

for i in range(len(phi)):
    plt.plot(x,phi[i]**2,label='Eigenvector %1i' %i)

plt.xlabel(r'MeV$^{-1}$')
plt.ylabel(r'$(4\pi r\psi)^2$')
plt.grid('on')
plt.legend()
plt.show()
```
